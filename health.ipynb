```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("./Health_heart_experimental.csv")
df.head(5)
```

```python
age = [21,30,35,39,61]
bmi = [21.0,21.0,20.0,33.0,32.0]

# create a line chart, age on x-axis, bmi on y-axis
plt.plot(age, bmi, color='green', marker='+', linestyle='dotted')

# add a title
plt.title("HEART HEALTH ")

# add a label to the y-axisage
plt.ylabel("Bmi")
plt.xlabel("age")
plt.show()
```

```python
age = [21,30,35,39,65]

bmi = [21,21,20,31,33]

# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [bmi]
plt.bar(range(len(age)), bmi)

plt.title("age of individuals")     # add a title
plt.ylabel("bmi of individuals")   # label the y-axis

# label x-axis with age at bar centers
plt.xticks(range(len(age)), age)

plt.show()
```
